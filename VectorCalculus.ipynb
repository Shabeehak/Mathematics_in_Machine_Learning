**Vector Calculus: Univariate, Bivariate, and Multivariate**

1. Univariate Calculus

```python
import sympy as sp

x = sp.Symbol('x')
f = x**3 - 6*x**2 + 9*x
df = sp.diff(f, x)  # First derivative
critical_points = sp.solve(df, x)

print("Derivative:", df)
print("Critical Points:", critical_points)
```

**Univariate Functions**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2 + 2*x + 1

# Derivative of the function
def f_derivative(x):
    return 2*x + 2

# Plot the function and its derivative
x = np.linspace(-10, 10, 100)
y = f(x)
y_prime = f_derivative(x)

plt.plot(x, y, label="f(x) = $x^2$ + 2x + 1")
plt.plot(x, y_prime, label="f'(x) = 2x + 2", linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Univariate Function and Its Derivative")
plt.show()
```

1. Derivatives (Rate of Change)

```python
from sympy import symbols, diff

x = symbols('x')
f = x**3 - 3*x**2 + 4*x + 1
derivative = diff(f, x)
print("Derivative of f(x):", derivative)
```

2. Applications of Derivatives
Critical Points: Where f'(x)=0. Used to find maxima, minima, and inflection points.

Maximizing profit or minimizing cost for a given function.

```python
from sympy import solve

f_prime = diff(f, x)
critical_points = solve(f_prime, x)
print("Critical Points:", critical_points)
```

3. Integration (Area Under the Curve)

```python
from sympy import integrate

# Indefinite integral
integral = integrate(f, x)
print("Indefinite Integral of f(x):", integral)
```

```python
# Definite integral
definite_integral = integrate(f, (x, 0, 2))  # Area from x=0 to x=2
print("Definite Integral of f(x) from 0 to 2:", definite_integral)
```

2. Bivariate Calculus

```python
import numpy as np

def gradient_f(x, y):
    df_dx = 2 * x  # f/x
    df_dy = 2 * y  # f/y
    return np.array([df_dx, df_dy])

x, y = 1, 2
grad = gradient_f(x, y)
print("Gradient at (1,2):", grad)
```

*Bivariate Functions*

```python
# Define symbols
x, y = symbols('x y')

# Define the function
f = x**2 + y**2

# Compute partial derivatives
partial_x = diff(f, x)
partial_y = diff(f, y)

print("f(x, y) =", f)
print("f/x =", partial_x)
print("f/y =", partial_y)
```

Gradient of a bivariate function

```python
import numpy as np

# Define the function f(x, y) = x^2 + y^2
def f(x, y):
    return x**2 + y**2

# Compute the gradient of f(x, y) with respect to x and y
def gradient_f(x, y):
    df_dx = 2 * x  # Partial derivative with respect to x
    df_dy = 2 * y  # Partial derivative with respect to y
    return np.array([df_dx, df_dy])

# Example point (x, y) = (1, 2)
x, y = 1, 2

# Compute the gradient at (1, 2)
grad = gradient_f(x, y)

print(f"The gradient of f(x, y) at (x, y) = ({x}, {y}) is: {grad}")
```

3. Multivariate

```python
import numpy as np

def gradient_f(x, y, z):
    df_dx = 2 * x
    df_dy = 2 * y
    df_dz = 2 * z
    return np.array([df_dx, df_dy, df_dz])

point = (1, 2, 3)
grad = gradient_f(*point)
print("Gradient at (1,2,3):", grad)
```

Gradient of a multivariable function using SymPy

```python
from sympy import symbols, Matrix, diff

x, y, z = symbols('x y z')
f = x**2 + y**2 + z**2
grad_f = Matrix([diff(f, var) for var in (x, y, z)])

print("Gradient:", grad_f)
```

Jacobian Matrix

```python
import sympy as sp

x, y = sp.symbols('x y')
F1 = x**2 + y
F2 = y**2 + x
F = sp.Matrix([F1, F2])

J = F.jacobian([x, y])
print("Jacobian Matrix:")
sp.pprint(J)
```
